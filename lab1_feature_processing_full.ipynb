## 1. Импорт библиотек
На данном этапе подключаются библиотеки для работы с данными и инструменты предварительной обработки из sklearn.

```
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler

np.random.seed(42)
```

## 2. Формирование датасета
Создается искусственный набор данных, содержащий числовые признаки (возраст и зарплата) и категориальные признаки (отдел и образование). После этого в данные специально добавляются пропуски.

```
n = 100

data = pd.DataFrame({
    'Age': np.random.randint(18, 65, n),
    'Salary': np.random.randint(30000, 150000, n),
    'Department': np.random.choice(['IT', 'HR', 'Finance', 'Marketing'], n),
    'Education': np.random.choice(['Bachelor', 'Master', 'PhD'], n)
})

for col in ['Age', 'Salary', 'Department']:
    idx = np.random.choice(data.index, size=10, replace=False)
    data.loc[idx, col] = np.nan

print('Первые строки датасета:')
display(data.head())

print('\nКоличество пропусков по столбцам:')
print(data.isnull().sum())
```

## 3. Обработка пропусков
Для числовых признаков пропуски заменяются средним значением признака. Для категориальных признаков используется наиболее часто встречающееся значение.

```
data_filled = data.copy()

num_imputer = SimpleImputer(strategy='mean')
data_filled[['Age', 'Salary']] = num_imputer.fit_transform(
    data_filled[['Age', 'Salary']]
)

cat_imputer = SimpleImputer(strategy='most_frequent')
data_filled[['Department']] = cat_imputer.fit_transform(
    data_filled[['Department']]
)

print('Количество пропусков после обработки:')
print(data_filled.isnull().sum())
```

## 4. Кодирование категориального признака методом Label Encoding
Каждой категории признака Education присваивается уникальный числовой код.

```
data_label = data_filled.copy()

le = LabelEncoder()
data_label['Education'] = le.fit_transform(data_label['Education'])

display(data_label.head())
```

## 5. Кодирование категориального признака методом One-Hot Encoding
Признак Department преобразуется в набор бинарных признаков.

```
data_ohe = pd.get_dummies(
    data_filled,
    columns=['Department'],
    dtype=int
)

display(data_ohe.head())
```

## 6. Нормализация методом Min-Max
Все числовые значения приводятся к диапазону от 0 до 1.

```
data_minmax = data_ohe.copy()

scaler = MinMaxScaler()

data_minmax[['Age', 'Salary']] = scaler.fit_transform(
    data_minmax[['Age', 'Salary']]
)

display(data_minmax.head())
```

## 7. Стандартизация данных
После стандартизации признаки имеют среднее значение около 0 и стандартное отклонение около 1.

```
data_standard = data_ohe.copy()

scaler_std = StandardScaler()

data_standard[['Age', 'Salary']] = scaler_std.fit_transform(
    data_standard[['Age', 'Salary']]
)

display(data_standard.head())
```

## 8. Вывод
Подводятся итоги выполнения лабораторной работы.

```
print('Вывод:')
print('1. Создан набор данных с числовыми и категориальными признаками.')
print('2. В данные добавлены пропуски.')
print('3. Выполнено заполнение пропусков.')
print('4. Выполнено кодирование категориальных признаков двумя способами.')
print('5. Выполнена нормализация и стандартизация числовых признаков.')
```
